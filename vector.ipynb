```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set()
```

Vector Addition if we have 2 vectors [2,3],[3,-2] then their addition will be [5,1] and subtraction will be [-1,5]. Visualizing them we get:

```python
plt.quiver(0, 0, 2, 3, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, 3, -2, angles='xy', scale_units='xy', scale=1, color='r') 

plt.xlim(-3, 6)
plt.ylim(-3, 6)
plt.show()
```

ADDITION

```python
plt.quiver(0, 0, 2, 3, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, 3, -2, angles='xy', scale_units='xy', scale=1, color='r') 
plt.quiver(0, 0, 5, 1, angles='xy', scale_units='xy', scale=1, color='b') #addition

plt.xlim(-3, 6)
plt.ylim(-3, 6)
plt.show()
```

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sum = vector_1+vector_2
print(sum)
```

Subtraction

```python
plt.quiver(0, 0, 2, 3, angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, 3, -2, angles='xy', scale_units='xy', scale=1, color='r') 
plt.quiver(0, 0, -1, 5, angles='xy', scale_units='xy', scale=1, color='g') #subtraction

plt.xlim(-3, 6)
plt.ylim(-3, 6)
plt.show()
```

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
subtraction = vector_1-vector_2
print(subtraction)
```

```python
v1= np.asarray([0,0,2,3])
result= 2*v1
print(result)
```
